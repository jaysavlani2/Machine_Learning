```python
x1 = df1.iloc[:,1].values
label=df1.iloc[:,0].values
```

```python
max=np.max(x1)
```

```python
explode=[]
for i in range(len(x1)):
  if(max==x1[i]):
    explode1=0.2
  else:
    explode1=0
  explode.append(explode1)
print(explode)
```

```python
plt.pie(x1,labels=label,explode=explode)
plt.show()
```
