**Analyse the cases of Confirmed, Death and Recovered figures from 01 March, 2021 in
Gujarat**

```python
import os
import glob
location='Data'
file_name=[]
for file in os.listdir(location):
  file_name.append(file)
print(file_name)
```

```python
combine_csv=pd.concat([pd.read_csv("Data/"+f) for f in file_name])
combine_csv.to_csv("combine_csv.csv",index=False,encoding='utf-8-sig')
```

```python
combine_csv
```

```python
gujarat = combine_csv[combine_csv['Province_State']=='Gujarat']
gujarat = gujarat.drop(['FIPS','Admin2','Province_State','Country_Region','Last_Update','Lat','Long_','Active','Combined_Key','Incident_Rate','Case_Fatality_Ratio'], axis = 1)
gujarat=gujarat.sort_values("Confirmed")
gujarat['Sr']=[i for i in range(len(gujarat))]
```

```python
gujarat
```

```python
plt.plot(gujarat['Sr'], gujarat['Confirmed'], label = "Confirmed", linestyle="-")
plt.plot(gujarat['Sr'], gujarat['Deaths'], label = "Deaths", linestyle="--")
plt.plot(gujarat['Sr'], gujarat['Recovered'], label = "Recovered", linestyle="-.")
plt.xticks(np.arange(0,gujarat.shape[0], 15))
plt.xlabel("Days")
plt.ylabel("Number of Cases")
plt.title("Gujarat Covid Cases")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures


# transforming the data to include another axis
x = gujarat['Sr'].values.reshape(-1, 1)
y = gujarat['Confirmed'].values.reshape(-1, 1)

polynomial_features= PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
```

```python
#64 -> 4th may
pre=np.array([64]).reshape(-1,1)
p = polynomial_features.fit_transform(pre)
g = model.predict(p)
guj_on_4 = g[0][0]-607422
guj_on_4
```

**Comparison of India and Gujarat**

```python
print("India on 4th May = ",int(On_4_may))
print("Gujarat on 4th May = ",int(guj_on_4))
print()
perct = (guj_on_4/On_4_may)*100
print("Gujarat Cases contribute",perct.round(2),"% of total cases in India")

plt.pie([perct.round(2),100-perct.round(2)],labels=['Gujarat','Rest of India'],
        explode=(0.1,0),shadow=True,autopct='%1.1f%%')
plt.title("Comparison")
plt.legend()
plt.show()
```

```python
import os
import glob
location='20 States'
file_name=[]
for file in os.listdir(location):
  file_name.append(file)
print(file_name)
```

```python
combine_csv=pd.concat([pd.read_csv("20 States/"+f) for f in file_name])
combine_csv.to_csv("combine_csv.csv",index=False,encoding='utf-8-sig')
```

```python
combine_csv
```

```python
States= combine_csv[combine_csv['Country_Region']=='India']
States = States.drop(['FIPS','Admin2','Lat','Country_Region','Long_','Deaths','Recovered','Active','Combined_Key','Incident_Rate','Case_Fatality_Ratio'], axis = 1)
```

```python
States
```

```python
new = States["Last_Update"].str.split(" ", n = 1, expand = True)
  
# making separate first name column from new data frame
States["Date"]= new[0]
  
# making separate last name column from new data frame
States["Time"]= new[1]
  
# Dropping old Name columns
States.drop(columns =["Time"], inplace = True)
States.drop(columns =["Last_Update"], inplace = True)
  
# df display
States
```

```python
January=States[States['Date']=='2021-01-02']
January=January.sort_values("Confirmed",ascending=False)
January=January.iloc[0:20,:]
print(January.shape)
January
```

```python
May=States[States['Date']=='2021-05-02']
May=May.sort_values("Confirmed",ascending=False)
May=May.iloc[0:20,:]
print(May.shape)
May
```
