```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#Binary Step Activation Function
def binaryStep(x):
  if x>=0:
    return 1
  else:
    return 0
```

```python
binaryStep(0)
```

```python
x=range(-10,11)
y=[]
for i in x:
  y.append(binaryStep(i))
plt.plot(x,y)
plt.title('Binary Step Activation Function')
plt.show()
```

```python
#Linear Activation Function
def Linear(x,m):
  return x*m
```

```python
x=range(-10,11)
y=[]
for i in x:
  y.append(Linear(i,1))
plt.plot(x,y)
plt.title('Linear Activation Function')
plt.show()
```

```python
#Sigmoid/Logistic Activation Function
def Sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
x=range(-10,11)
y=[]
for i in x:
  y.append(Sigmoid(i))
plt.plot(x,y)
plt.title('Sigmoid Activation Function')
plt.show()
```

```python
#Tanh Activation Function
def Tanh(x):
  return (1-np.exp(-2*x))/(1+np.exp(-2*x))
```

```python
x=range(-10,11)
y=[]
for i in x:
  y.append(Tanh(i))
plt.plot(x,y)
plt.title('Hyperbolic Activation Function')
plt.show()
```

```python
#ReLU Activation Function
def ReLU(x):
  if x<0:
    return(0)
  else:
    return(x)
```

```python
x=range(-10,11)
y=[]
for i in x:
  y.append(ReLU(i))
plt.plot(x,y)
plt.title('Rectified Linear unit Activation Function')
plt.show()
```

```python
#Leaky ReLU Activation Function
def Leaky_ReLU(x):
  if x>0 :
    return x
  else :
    return 0.01*x
```

```python
x=range(-10,11)
y=[]
for i in x:
  y.append(Leaky_ReLU(i))
plt.plot(x,y)
plt.title('Leaky Rectified Linear unit Activation Function')
plt.show()
```
