# Gujarat State Analysis

```python
date_india = pd.DataFrame()
```

```python
import os

for i in os.listdir("data"):
    
    temp_date=pd.read_csv(i)  
    temp_date_i = temp_date[temp_date['Country_Region']=="India"]

    temp_date_ind = pd.DataFrame()
    temp_date_ind['State'] = temp_date_i['Province_State']
    temp_date_ind['Confirmed'] = temp_date_i['Confirmed']
    temp_date_ind['Deaths'] = temp_date_i['Deaths']
    temp_date_ind['Recovered'] = temp_date_i['Recovered']
    d=i[:-4]
    temp_date_ind['Date']=d
    
    date_india = date_india.append(temp_date_ind)  

date_india.head()
```

```python
date_india['Date'].unique()
```

```python
gujarat_data = date_india[date_india['State']=="Gujarat"]
gujarat_data.reset_index(inplace = True)
gujarat_data.head()
```

```python
plt.plot(gujarat_data['Date'], gujarat_data['Confirmed'], label = "Confirmed", linestyle="-")
plt.plot(gujarat_data['Date'], gujarat_data['Deaths'], label = "Deaths", linestyle="--")
plt.plot(gujarat_data['Date'], gujarat_data['Recovered'], label = "Recovered", linestyle="-.")
plt.xticks(np.arange(0,gujarat_data.shape[0], 15))
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.title("Gujarat Covid Cases")
plt.legend()
plt.show()
```

# Making LR Model For Predicting Gujarat Cases


```python
data = gujarat_data
data["day_seq"]=[i for i in range(data.shape[0])]
data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures


# transforming the data to include another axis
x = data['day_seq'].values.reshape(-1, 1)
y = data['Confirmed'].values.reshape(-1, 1)

polynomial_features= PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
```

```python
#64 -> 4th may
pre=np.array([64]).reshape(-1,1)
p = polynomial_features.fit_transform(pre)
g = model.predict(p)
guj_on_4 = g[0][0]-594602
guj_on_4
```

# Comapring Guj vs India covid cases on 4th May

```python
print("Total predicted cases in India on 4th May = ",int(india_on_4))
print("Total predicted cases in Gujarat on 4th May = ",int(guj_on_4))
print()
perct = (guj_on_4*100)/india_on_4
print("Gujarat Cases contribute",perct.round(2),"% of total cases in India")

plt.pie([perct.round(2),100-perct.round(2)],labels=['Gujarat','Rest of India'],
        explode=(0.1,0),shadow=True,autopct='%1.1f%%')
plt.title("Covid cases on 4th May")
plt.legend()
plt.show()
```
