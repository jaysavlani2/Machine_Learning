```python
email_data = pd.read_csv("spam_ham_dataset.csv")
```

```python
email_data.head()
```

```python
email_data.isnull().sum()
```

```python
# dropping unwanted columns
email_data = email_data.drop("Unnamed: 0", axis = 1)
email_data = email_data.drop("label_num", axis = 1)
```

```python
email_data['type'] = email_data.label.map({'ham': 1 , 'spam' : 0}) 
```

```python
email_data.head()
```

```python
X=email_data.iloc[:,1].values
Y=email_data.iloc[:,2].values
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection  import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# vectorising the text
vect = CountVectorizer(stop_words='english')
```

```python
vect.fit(X_train)
```

```python
vect.vocabulary_
```

```python
vect.get_feature_names()
```

```python
# transform
X_train_transformed = vect.transform(X_train)
X_test_tranformed =vect.transform(X_test)
```

```python
print(X_train_transformed)
```

```python
from sklearn.naive_bayes import BernoulliNB
# instantiate bernoulli NB object
bnb = BernoulliNB()

# fit 
bnb.fit(X_train_transformed,Y_train)

# predict class
y_pred_class = bnb.predict(X_test_tranformed)

# predict probability
y_pred_proba =bnb.predict_proba(X_test_tranformed)

# accuracy
from sklearn import metrics
metrics.accuracy_score(Y_test, y_pred_class)
```

```python
metrics.confusion_matrix(Y_test, y_pred_class)
```
